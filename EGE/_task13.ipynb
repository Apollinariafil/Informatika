В терминологии сетей TCP/IP маской сети называют двоичное число, которое показывает, какая часть IP-адреса узла сети относится к адресу сети, а какая — к адресу узла в этой сети. Адрес сети получается в результате применения поразрядной конъюнкции к заданному адресу узла и маске сети.

Сеть задана IP-адресом 123.222.0.192 и маской сети 255.255.255.224. Сколько в этой сети IP-адресов, в которых количество нулей и единиц в двоичной записи IP-адреса одинаково? В ответе укажите только число.

```python
from ipaddress import *  # Импортируем только нужную функцию

# Создаем объект сети с указанным адресом и маской
net1 = ip_network('123.222.0.192/255.255.255.224', strict=True)
ip_a = ip_address('123.222.0.192')

print(ip_a)
print(bin(int(ip_a))[2:])
# Выводим все хосты в сети
# print(*net1.hosts())

# Инициализируем счетчик для подсчета хостов с равным количеством единиц и нулей в бинарном представлении
k = 0

# Перебираем все хосты в сети
for x in net1.hosts():
    # Преобразуем IP-адрес в 32-битное бинарное представление
    n = bin(int(x))[2:]  # Используем zfill для заполнения нулями до 32 бит
    # Проверяем, равно ли количество единиц количеству нулей
    if n.count('1') == n.count('0'):
        k += 1  # Увеличиваем счетчик, если условие выполняется

# Выводим результат
print(k)
```

Сеть задана IP-адресом 172.16.168.0 и маской сети 255.255.248.0.

Сколько в этой сети IP-адресов, для которых количество единиц в двоичной записи IP-адреса не кратно 5?

В ответе укажите только число.

```python
from ipaddress import *  

network = ip_network('172.16.168.0/255.255.255.0', strict=False)

ip_a = ip_address('172.16.168.0')
ip_m = ip_address('255.255.248.0')

print(ip_a)
print(bin(int(ip_a))[2:])
print(bin(int(ip_m))[2:])

# Счетчик для адресов, где количество единиц не кратно 5
count = 0

# Перебираем все хосты в сети
# Метод hosts() возвращает итератор всех хостов в сети (исключая сетевой и широковещательный адреса).
print(network.network_address)  # адрес сети
print(network.broadcast_address)  # широковещательный адрес.

for ip in network.hosts():
    # Преобразуем IP-адрес в двоичную строку
    binary_ip = bin(int(ip))[2:].zfill(32)
    # Считаем количество единиц
    ones_count = binary_ip.count('1')
    # Проверяем, не кратно ли 5
    if ones_count % 5 != 0:
        count += 1

# Выводим результат
print(count)
```
