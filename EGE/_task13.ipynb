В терминологии сетей TCP/IP маской сети называют двоичное число, которое показывает, какая часть IP-адреса узла сети относится к адресу сети, а какая — к адресу узла в этой сети. Адрес сети получается в результате применения поразрядной конъюнкции к заданному адресу узла и маске сети.

Сеть задана IP-адресом 172.16.168.0 и маской сети 255.255.248.0.

Сколько в этой сети IP-адресов, для которых количество единиц в двоичной записи IP-адреса не кратно 5?

В ответе укажите только число.

````python
from ipaddress import *  

network = ip_network('172.16.168.0/255.255.248.0', strict=False)

ip_a = ip_address('172.16.168.0')
ip_m = ip_address('255.255.248.0')

print(ip_a)
print(bin(int(ip_a))[2:])
print(bin(int(ip_m))[2:])

# Счетчик для адресов, где количество единиц не кратно 5
count = 0

# Перебираем все хосты в сети
# Метод hosts() возвращает итератор всех хостов в сети (исключая сетевой и широковещательный адреса).
# print(network.network_address)  # адрес сети 172.16.168.0
# print(network.broadcast_address)  # широковещательный адрес 172.16.175.255


for ip in network.hosts():
    # Преобразуем IP-адрес в двоичную строку
    binary_ip = bin(int(ip))[2:].zfill(32)
    # Считаем количество единиц
    ones_count = binary_ip.count('1')
    # Проверяем, не кратно ли 5
    if ones_count % 5 != 0:
        count += 1

network_address = bin(int(network.network_address))[2:].zfill(32)
broadcast_address = bin(int(network.broadcast_address))[2:].zfill(32)

if network_address.count('1') % 5 != 0:
    count += 1
if broadcast_address.count('1') % 5 != 0:
    count += 1

# Выводим результат
print(count)
````

Два узла, находящиеся в одной сети, имеют IP-адреса 121.171.5.70 и 121.171.5.107. Укажите наименьшее возможное количество адресов в этой сети.


Чтобы решить задачу, нужно определить, какая маска сети позволяет объединить два заданных IP-адреса в одну сеть с минимальным количеством адресов. Для этого выполним следующие шаги:

---

### Шаги решения:
1. **Переведем IP-адреса в двоичный вид**:
   - \(121.171.5.70\) → `01111001.10101011.00000101.01000110`
   - \(121.171.5.107\) → `01111001.10101011.00000101.01101011`

2. **Найдем общую часть адресов**:
   - Сравниваем двоичные представления адресов по битам слева направо.
   - Общая часть заканчивается на 26-м бите:
     ```
     01111001.10101011.00000101.01000110 (121.171.5.70)
     01111001.10101011.00000101.01101011 (121.171.5.107)
     ```
     Общая часть: `01111001.10101011.00000101.01` (26 бит).

3. **Определим маску сети**:
   - Маска сети должна быть такой, чтобы первые 26 бит были единицами, а остальные — нулями.
   - Маска: `255.255.255.192` (в двоичном виде: `11111111.11111111.11111111.11000000`).

4. **Вычислим количество ХОСТОВ в сети**:
   - Количество хостов в сети определяется формулой: \(2^{32 - \text{префикс}} - 2\).
   - Префикс: 26.
   - Количество адресов: \(2^{32-26} - 2 = 2^6 - 2 = 64 - 2 = 62\).

---

### Результат:
Наименьшее возможное количество адресов в этой сети — **62 + 2 = 64**.

---

### Проверка:
1. **Адрес сети**:
   - Применим поразрядную конъюнкцию к одному из IP-адресов и маске:
     ```
     121.171.5.70 & 255.255.255.192 = 121.171.5.64
     ```
   - Адрес сети: \(121.171.5.64\).

2. **Диапазон адресов**:
   - Начальный адрес: \(121.171.5.64\).
   - Конечный адрес: \(121.171.5.127\).
   - Количество адресов: \(127 - 64 + 1 = 64\) (включая сетевой и широковещательный адреса).
   - Количество хостов: \(64 - 2 = 62\).

---

**Ответ:** `64`.

Адрес сети получается в результате применения поразрядной конъюнкции к заданному адресу узла и маске сети. Сеть задана IP-адресом 192.168.32.160 и маской сети 255.255.255.240.

В ответе укажите только число.

````python
from ipaddress import *  

network = ip_network('192.168.32.160/255.255.255.240', strict=False)


# Счетчик для адресов, где количество единиц не кратно 5
count = 0


for ip in network.hosts():
    # Преобразуем IP-адрес в двоичную строку
    binary_ip = bin(int(ip))[2:].zfill(32)
    # Считаем количество единиц
    ones_count = binary_ip.count('0')
    # Проверяем, не кратно ли 5
    if ones_count > 21:
        count += 1

network_address = bin(int(network.network_address))[2:].zfill(32)
broadcast_address = bin(int(network.broadcast_address))[2:].zfill(32)

if network_address.count('0') > 21:
    count += 1
if broadcast_address.count('0') > 21:
    count += 1

# Выводим результат
print(count)
````
