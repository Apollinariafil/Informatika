```python
""" Автомат получает на вход трёхзначное число. По этому числу строится новое число по следующим правилам.
1.Складываются первая и вторая, а также вторая и третья цифры исходного числа.
2.Полученные два числа записываются друг за другом в порядке убывания (без разделителей).

Пример. Исходное число: 348. Суммы: 3+4=7; 4+8=12. 

Результат: 127. 

Укажите наименьшее число, в результате обработки которого автомат выдаст число 1412. 
https://inf-ege.sdamgia.ru/problem?id=7663
"""
# Вариант решения c использованием функции


def myresult(i):
    stroka = str(i)
    num1 = int(stroka[0])+int(stroka[1])
    num2 = int(stroka[1])+int(stroka[2])
    first = max(num1, num2)
    second = min(num1, num2)
    result = str(first) + str(second)
    if result == '1412':
        return True

    return False


# Можно наполнить список всеми подходящими и найти минимум
nums = []

for i in range(100, 1000):
    if myresult(i):
        nums.append(i)
print(min(nums))


# # Вариант решения без использования функции
# for i in range(100,1000):

#     stroka = str(i)
#     num1 = int(stroka[0])+int(stroka[1])
#     num2 = int(stroka[1])+int(stroka[2])
#     first = max(num1, num2)
#     second = min(num1, num2)
#     result = str(first) + str(second)

#     if result == '1412':
#         nums.append(i)
# print(nums)
# print(sorted(nums))
```

```python
# На вход алгоритма подаётся натуральное число N. Алгоритм строит по нему новое число R следующим образом.
# 1.  Строится двоичная запись числа N.
# 2.  К этой записи дописываются справа ещё несколько разрядов по следующему правилу:
# а) если N чётное, то к нему справа приписываются два нуля, а слева единица;
# б) если N нечётное, то к нему справа приписывается в двоичном виде сумма цифр его двоичной записи;
# Полученная таким образом запись (в ней как минимум на один разряд больше, чем в записи исходного числа N) является двоичной записью искомого числа R.
# Например, исходное число 4(10)=100(2) преобразуется в число 110000(2)=48101100002=48(10),
# Укажите такое число N, для которого число R является наименьшим среди чисел, превышающих 190. В ответе это число запишите в десятичной системе счисления

def myFunc(N):
    # получаем двоичную запись числа
    b = str(bin(N)[2:])
    if N % 2 == 0:
        b ='1' + b + '00'
    else:
        sum = 0
        for j in b:
            sum += int(j)
        b = b + str(bin(sum)[2:])

    b = int(b, 2)

    return b

res = []
for i in range(1, 1000):
    R = myFunc(i)
    if R > 190:
        res.append(R)
        print(f'i = {i} R = {R}')
    # if R == 192:
    #     print(f'i = {i} R = {R}')




print(min(res))
```
