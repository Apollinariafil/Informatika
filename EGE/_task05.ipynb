```python
""" Шастин Л.
На вход алгоритма подаётся нечётное натуральное число N. Алгоритм строит по нему новое число R следующим образом.

Строится четверичная запись числа N.
Далее эта запись обрабатывается по следующему правилу:
а) если число N делится на 3, то первая и последняя цифры меняются местами, а затем в конец дописывается единица;
б) если число N на 3 не делится, то остаток от деления на 3 дописывается в конец числа.
Полученная таким образом запись является четверичной записью искомого числа R.
Результат переводится в десятичную систему и выводится на экран.
Например, для исходного числа 11 = 23(4)
результатом является число 232(4) = 46, а для исходного числа 13 = 31(4)
это число 311(4) = 53
Укажите максимальное число R, не превышающее 340, которое может быть получено с помощью описанного алгоритма.

В ответе запишите это число в десятичной системе счисления """

# функция перевода в 4-ю систему исчисления


def to_4(n):
    s = ""
    # будем делать цикл пока n больше нуля...
    while (n > 0):
        # считаем остаток от деления на 4 (в случае 7-чной системы на 7)
        ostatok = n % 4
        # а к самой n применяем деление без остатка
        n = n // 4
        # делаем новую строку из остаков от деления
        s += str(ostatok)

    return s[::-1]


def F(n):
    n_4 = to_4(n)
    s = str(n_4)  # делаем строку из числа
    r = ''
    if n % 3 == 0:
        if len(s) > 1:
            r = s[-1]+s[1:-1]+s[0]+'1'
        else:
            r = s+'1'
    else:
            r = s + str(n % 3)

    return int(r, 4)

res = []
for i in range(1, 100000, 2):
    if F(i)<=340:
        res.append(F(i))
        # print(f'i = {i} F(i) = {F(i)}')

print(sorted(res, reverse=True))
print(max(res))

# print(f'i = {i} F(i) = {F(i)} 10 = {int(F(i), 4)}')
```

```python
""" Автомат получает на вход трёхзначное число. По этому числу строится новое число по следующим правилам.
1.Складываются первая и вторая, а также вторая и третья цифры исходного числа.
2.Полученные два числа записываются друг за другом в порядке убывания (без разделителей).

Пример. Исходное число: 348. Суммы: 3+4=7; 4+8=12. 

Результат: 127. 

Укажите наименьшее число, в результате обработки которого автомат выдаст число 1412. 
https://inf-ege.sdamgia.ru/problem?id=7663
"""
# Вариант решения c использованием функции


def myresult(i):
    stroka = str(i)
    num1 = int(stroka[0])+int(stroka[1])
    num2 = int(stroka[1])+int(stroka[2])
    first = max(num1, num2)
    second = min(num1, num2)
    result = str(first) + str(second)
    if result == '1412':
        return True

    return False


# Можно наполнить список всеми подходящими и найти минимум
nums = []

for i in range(100, 1000):
    if myresult(i):
        nums.append(i)
print(min(nums))


# # Вариант решения без использования функции
# for i in range(100,1000):

#     stroka = str(i)
#     num1 = int(stroka[0])+int(stroka[1])
#     num2 = int(stroka[1])+int(stroka[2])
#     first = max(num1, num2)
#     second = min(num1, num2)
#     result = str(first) + str(second)

#     if result == '1412':
#         nums.append(i)
# print(nums)
# print(sorted(nums))
```

```python
# На вход алгоритма подаётся натуральное число N. Алгоритм строит по нему новое число R следующим образом.
# 1.  Строится двоичная запись числа N.
# 2.  К этой записи дописываются справа ещё несколько разрядов по следующему правилу:
# а) если N чётное, то к нему справа приписываются два нуля, а слева единица;
# б) если N нечётное, то к нему справа приписывается в двоичном виде сумма цифр его двоичной записи;
# Полученная таким образом запись (в ней как минимум на один разряд больше, чем в записи исходного числа N) является двоичной записью искомого числа R.
# Например, исходное число 4(10)=100(2) преобразуется в число 110000(2)=48101100002=48(10),
# Укажите такое число N, для которого число R является наименьшим среди чисел, превышающих 190. В ответе это число запишите в десятичной системе счисления

def myFunc(N):
    # получаем двоичную запись числа
    b = str(bin(N)[2:])
    if N % 2 == 0:
        b ='1' + b + '00'
    else:
        sum = 0
        for j in b:
            sum += int(j)
        b = b + str(bin(sum)[2:])

    b = int(b, 2)

    return b

res = []

for i in range(1, 10000):
    R = myFunc(i)
    if R > 190:
        # print(f'i = {i} R = {R}')
        res.append(R)

print(min(res))
```
