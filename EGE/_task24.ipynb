````python
# –∏—Å—Ö–æ–¥–Ω—ã–π –∫–æ–¥ –ö–≠–û

text_string = open('files/24.txt').readline()
split_text = text_string.split('A')

max_value = -float('inf')

for a, b, c in zip(split_text, split_text[1:], split_text[2:]):
    if a == b == c:
        max_value = max(max_value,len(c*3))


res = max_value + 4
print(res)
````

- –ß–∏—Å–ª–æ `4` –¥–æ–±–∞–≤–ª–µ–Ω–æ, –≤–µ—Ä–æ—è—Ç–Ω–æ, –¥–ª—è —É—á—ë—Ç–∞ —Å–∏–º–≤–æ–ª–æ–≤ `'A'`, –∫–æ—Ç–æ—Ä—ã–µ –±—ã–ª–∏ —É–¥–∞–ª–µ–Ω—ã –ø—Ä–∏ —Ä–∞–∑–±–∏–µ–Ω–∏–∏ —Å—Ç—Ä–æ–∫–∏.

- –í—ã–≤–æ–¥–∏—Ç –∏—Ç–æ–≥–æ–≤–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ `res` –Ω–∞ —ç–∫—Ä–∞–Ω.

---

### –ß—Ç–æ –¥–µ–ª–∞–µ—Ç —ç—Ç–æ—Ç –∫–æ–¥?
–ö–æ–¥ –∞–Ω–∞–ª–∏–∑–∏—Ä—É–µ—Ç —Å—Ç—Ä–æ–∫—É –∏–∑ —Ñ–∞–π–ª–∞, —Ä–∞–∑–±–∏–≤–∞–µ—Ç –µ—ë –ø–æ —Å–∏–º–≤–æ–ª—É `'A'`, –∏—â–µ—Ç —Ç—Ä–∏ –ø–æ–¥—Ä—è–¥ –∏–¥—É—â–∏—Ö –æ–¥–∏–Ω–∞–∫–æ–≤—ã—Ö —Ñ—Ä–∞–≥–º–µ–Ω—Ç–∞ –∏ –≤—ã—á–∏—Å–ª—è–µ—Ç –º–∞–∫—Å–∏–º–∞–ª—å–Ω—É—é –¥–ª–∏–Ω—É —Ç–∞–∫–∏—Ö —Ñ—Ä–∞–≥–º–µ–Ω—Ç–æ–≤, —É–≤–µ–ª–∏—á–µ–Ω–Ω—É—é –Ω–∞ 4.

---

–ï—Å–ª–∏ —É –≤–∞—Å –µ—Å—Ç—å –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –≤–æ–ø—Ä–æ—Å—ã –∏–ª–∏ –Ω—É–∂–Ω–æ —á—Ç–æ-—Ç–æ —É—Ç–æ—á–Ω–∏—Ç—å, –¥–∞–π—Ç–µ –∑–Ω–∞—Ç—å! üòä

````python

# –î–æ—Å—Ç—É–ø–µ–Ω —Ñ–∞–π–ª –¥–ª—è —á—Ç–µ–Ω–∏—è: 24.txt

s = open('24.txt').readline()
sg = 'CDF'
gl = 'AO'
mx = 0

cnt, i, j = 0, 0, 0
while j < len(s):
    if cnt <= 5:
        mx = max(mx, j-i+1)
        if j + 1 < len(s):
            if s[j] in sg and s[j+1] in gl:
                cnt += 1
        j += 1
    if s[i] in sg and s[i + 1] in gl:
        cnt -= 1
    i += 1

print(mx)
````

### –ß—Ç–æ –¥–µ–ª–∞–µ—Ç —ç—Ç–æ—Ç –∫–æ–¥?
–ö–æ–¥ –∏—â–µ—Ç –≤ —Å—Ç—Ä–æ–∫–µ `s` –ø–æ–¥—Å—Ç—Ä–æ–∫—É –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–π –¥–ª–∏–Ω—ã, –≤ –∫–æ—Ç–æ—Ä–æ–π –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–∞—Ä "—Å–æ–≥–ª–∞—Å–Ω–∞—è + –≥–ª–∞—Å–Ω–∞—è" –Ω–µ –ø—Ä–µ–≤—ã—à–∞–µ—Ç 5. –î–ª—è —ç—Ç–æ–≥–æ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –º–µ—Ç–æ–¥ "—Å–∫–æ–ª—å–∑—è—â–µ–≥–æ –æ–∫–Ω–∞":
- –û–∫–Ω–æ —Ä–∞—Å—à–∏—Ä—è–µ—Ç—Å—è –≤–ø—Ä–∞–≤–æ (`j` —É–≤–µ–ª–∏—á–∏–≤–∞–µ—Ç—Å—è), –µ—Å–ª–∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–∞—Ä "—Å–æ–≥–ª–∞—Å–Ω–∞—è + –≥–ª–∞—Å–Ω–∞—è" –Ω–µ –ø—Ä–µ–≤—ã—à–∞–µ—Ç 5.
- –û–∫–Ω–æ —Å–¥–≤–∏–≥–∞–µ—Ç—Å—è –≤–ø—Ä–∞–≤–æ (`i` —É–≤–µ–ª–∏—á–∏–≤–∞–µ—Ç—Å—è), –µ—Å–ª–∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–∞—Ä –ø—Ä–µ–≤—ã—à–∞–µ—Ç 5.

````python
# –û—Ç–∫—Ä—ã–≤–∞–µ–º —Ñ–∞–π–ª –∏ —á–∏—Ç–∞–µ–º –ø–µ—Ä–≤—É—é —Å—Ç—Ä–æ–∫—É
with open('24.txt') as file:
    line = file.readline()

# –û–ø—Ä–µ–¥–µ–ª—è–µ–º —Å–æ–≥–ª–∞—Å–Ω—ã–µ –∏ –≥–ª–∞—Å–Ω—ã–µ –±—É–∫–≤—ã
consonants = 'CDF'  # –°–æ–≥–ª–∞—Å–Ω—ã–µ
vowels = 'AO'       # –ì–ª–∞—Å–Ω—ã–µ

# –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ
max_length = 0      # –ú–∞–∫—Å–∏–º–∞–ª—å–Ω–∞—è –¥–ª–∏–Ω–∞ –ø–æ–¥—Ö–æ–¥—è—â–µ–π –ø–æ–¥—Å—Ç—Ä–æ–∫–∏
pair_count = 0      # –°—á—ë—Ç—á–∏–∫ –ø–∞—Ä "—Å–æ–≥–ª–∞—Å–Ω–∞—è + –≥–ª–∞—Å–Ω–∞—è"
left = 0            # –õ–µ–≤–∞—è –≥—Ä–∞–Ω–∏—Ü–∞ –æ–∫–Ω–∞
right = 0           # –ü—Ä–∞–≤–∞—è –≥—Ä–∞–Ω–∏—Ü–∞ –æ–∫–Ω–∞

# –û—Å–Ω–æ–≤–Ω–æ–π —Ü–∏–∫–ª: –ø—Ä–æ—Ö–æ–¥–∏–º –ø–æ —Å—Ç—Ä–æ–∫–µ —Å –ø–æ–º–æ—â—å—é "—Å–∫–æ–ª—å–∑—è—â–µ–≥–æ –æ–∫–Ω–∞"
while right < len(line):
    # –ï—Å–ª–∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–∞—Ä "—Å–æ–≥–ª–∞—Å–Ω–∞—è + –≥–ª–∞—Å–Ω–∞—è" –Ω–µ –ø—Ä–µ–≤—ã—à–∞–µ—Ç 5
    if pair_count <= 5:
        # –û–±–Ω–æ–≤–ª—è–µ–º –º–∞–∫—Å–∏–º–∞–ª—å–Ω—É—é –¥–ª–∏–Ω—É –ø–æ–¥—Ö–æ–¥—è—â–µ–π –ø–æ–¥—Å—Ç—Ä–æ–∫–∏
        current_length = right - left + 1
        max_length = max(max_length, current_length)

        # –ü—Ä–æ–≤–µ—Ä—è–µ–º, –µ—Å—Ç—å –ª–∏ –ø–∞—Ä–∞ "—Å–æ–≥–ª–∞—Å–Ω–∞—è + –≥–ª–∞—Å–Ω–∞—è" –Ω–∞ —Ç–µ–∫—É—â–µ–π –ø–æ–∑–∏—Ü–∏–∏
        if right + 1 < len(line) and line[right] in consonants and line[right + 1] in vowels:
            pair_count += 1  # –£–≤–µ–ª–∏—á–∏–≤–∞–µ–º —Å—á—ë—Ç—á–∏–∫ –ø–∞—Ä

        right += 1  # –†–∞—Å—à–∏—Ä—è–µ–º –æ–∫–Ω–æ –≤–ø—Ä–∞–≤–æ

    # –ï—Å–ª–∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–∞—Ä –ø—Ä–µ–≤—ã—à–∞–µ—Ç 5, —Å–¥–≤–∏–≥–∞–µ–º –ª–µ–≤—É—é –≥—Ä–∞–Ω–∏—Ü—É –æ–∫–Ω–∞
    if pair_count > 5:
        # –ï—Å–ª–∏ –ª–µ–≤–∞—è –≥—Ä–∞–Ω–∏—Ü–∞ –æ–∫–Ω–∞ —É–±–∏—Ä–∞–µ—Ç –ø–∞—Ä—É "—Å–æ–≥–ª–∞—Å–Ω–∞—è + –≥–ª–∞—Å–Ω–∞—è", —É–º–µ–Ω—å—à–∞–µ–º —Å—á—ë—Ç—á–∏–∫
        if left + 1 < len(line) and line[left] in consonants and line[left + 1] in vowels:
            pair_count -= 1

        left += 1  # –°–¥–≤–∏–≥–∞–µ–º –æ–∫–Ω–æ –≤–ø—Ä–∞–≤–æ

# –í—ã–≤–æ–¥–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç
print(max_length)
````

–ö–æ–Ω–µ—á–Ω–æ! –Ø —É–ª—É—á—à—É –≤–∞—à –∫–æ–¥, —Å–¥–µ–ª–∞–≤ –µ–≥–æ –±–æ–ª–µ–µ —á–∏—Ç–∞–µ–º—ã–º –∏ –ø–æ–Ω—è—Ç–Ω—ã–º. –û—Å–Ω–æ–≤–Ω—ã–µ –∏–∑–º–µ–Ω–µ–Ω–∏—è:

# –û—Ç–∫—Ä—ã–≤–∞–µ–º —Ñ–∞–π–ª –∏ —á–∏—Ç–∞–µ–º –ø–µ—Ä–≤—É—é —Å—Ç—Ä–æ–∫—É

````python
# –í —Ç–µ–∫—Å—Ç–æ–≤–æ–º —Ñ–∞–π–ª–µ —Ä–∞–∑–º–µ—Ä–æ–º  —Å–∏–º–≤–æ–ª–æ–≤ ‚Äî —Ç–æ–ª—å–∫–æ –∑–∞–≥–ª–∞–≤–Ω—ã–µ –ª–∞—Ç–∏–Ω—Å–∫–∏–µ –±—É–∫–≤—ã D, R, A, G, O, N. 
# –û–ø—Ä–µ–¥–µ–ª–∏—Ç–µ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—É—é –¥–ª–∏–Ω—É –ø–æ–¥–ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏, 
# –≤ –∫–æ—Ç–æ—Ä–æ–π –∫–∞–∂–¥–∞—è –∏–∑ –≥–ª–∞—Å–Ω—ã—Ö –±—É–∫–≤ (–ê, –û) –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è –Ω–µ –±–æ–ª–µ–µ 100 —Ä–∞–∑.

# –û—Ç–∫—Ä—ã–≤–∞–µ–º —Ñ–∞–π–ª –∏ —á–∏—Ç–∞–µ–º –ø–µ—Ä–≤—É—é —Å—Ç—Ä–æ–∫—É
with open('files/24 _003.txt') as file:
    text_string = file.readline()

# text_string = 'DDDRRRAAAGGGGO–û–û–ûDDDRRRAAAGGGGO–û–û–ûDDDRRRAAAGGGGO–û–û–ûDDDRRRAAAGGGGO–û–û–ûDDDRRRAAAGGGGO–û–û–ûDDDRRRAAAGGGGO–û–û–û'

# –û–ø—Ä–µ–¥–µ–ª—è–µ–º —Å–æ–≥–ª–∞—Å–Ω—ã–µ –∏ –≥–ª–∞—Å–Ω—ã–µ –±—É–∫–≤—ã
count_A = 0
count_O = 0

# –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ
max_length = 0      # –ú–∞–∫—Å–∏–º–∞–ª—å–Ω–∞—è –¥–ª–∏–Ω–∞ –ø–æ–¥—Ö–æ–¥—è—â–µ–π –ø–æ–¥—Å—Ç—Ä–æ–∫–∏
left = 0            # –õ–µ–≤–∞—è –≥—Ä–∞–Ω–∏—Ü–∞ –æ–∫–Ω–∞
right = 0           # –ü—Ä–∞–≤–∞—è –≥—Ä–∞–Ω–∏—Ü–∞ –æ–∫–Ω–∞

# –û—Å–Ω–æ–≤–Ω–æ–π —Ü–∏–∫–ª: –ø—Ä–æ—Ö–æ–¥–∏–º –ø–æ —Å—Ç—Ä–æ–∫–µ —Å –ø–æ–º–æ—â—å—é "—Å–∫–æ–ª—å–∑—è—â–µ–≥–æ –æ–∫–Ω–∞"
while right < len(text_string):

    if count_A <= 100 and count_O <= 100:

        # –û–±–Ω–æ–≤–ª—è–µ–º –º–∞–∫—Å–∏–º–∞–ª—å–Ω—É—é –¥–ª–∏–Ω—É –ø–æ–¥—Ö–æ–¥—è—â–µ–π –ø–æ–¥—Å—Ç—Ä–æ–∫–∏
        current_length = right - left
        max_length = max(max_length, current_length)


        # –£–≤–µ–ª–∏—á–∏–≤–∞–µ–º —Å—á–µ—Ç—á–∏–∫–∏ –¥–ª—è –≥–ª–∞—Å–Ω—ã—Ö A –∏ O
        if text_string[right] == 'A':
            count_A += 1
        elif text_string[right] == 'O':
            count_O += 1

        right += 1  # –†–∞—Å—à–∏—Ä—è–µ–º –æ–∫–Ω–æ –≤–ø—Ä–∞–≤–æ

    else: 
        # –ï—Å–ª–∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ A –∏–ª–∏ O –ø—Ä–µ–≤—ã—à–∞–µ—Ç 100, –¥–≤–∏–≥–∞–µ–º left
        if text_string[left] == 'A':
            count_A -= 1
        elif text_string[left] == 'O':
            count_O -= 1
        left += 1  # –°–¥–≤–∏–≥–∞–µ–º –æ–∫–Ω–æ –≤–ø—Ä–∞–≤–æ



# –í—ã–≤–æ–¥–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç
print(max_length)
````
