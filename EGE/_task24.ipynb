–¢–µ–∫—Å—Ç–æ–≤—ã–π —Ñ–∞–π–ª —Å–æ—Å—Ç–æ–∏—Ç –Ω–µ –±–æ–ª–µ–µ, —á–µ–º –∏–∑ 10‚Å∂ —Å–∏–º–≤–æ–ª–æ–≤ –∏–∑ –Ω–∞–±–æ—Ä–∞ A, B, C, D, E, F.

–ù–∞–π–¥–∏—Ç–µ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—É—é –¥–ª–∏–Ω—É —Å—Ç—Ä–æ–∫–∏ –≤–∏–¥–∞ –ê*–ê*–ê*–ê, –≥–¥–µ –º–µ–∂–¥—É –±—É–∫–≤–∞–º–∏ –ê —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω—ã –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ –≥—Ä—É–ø–ø—ã —Å–∏–º–≤–æ–ª–æ–≤, –Ω–µ —Å–æ–¥–µ—Ä–∂–∞—â–∏–µ –±—É–∫–≤ –ê. –ù–∞–ø—Ä–∏–º–µ—Ä, –≤ —Å—Ç—Ä–æ–∫–µ BDADBADBADBABDAFABDA —Ç–∞–∫–∞—è –ø–æ–¥—Å—Ç—Ä–æ–∫–∞ ADBADBADBA (–¥–ª–∏–Ω–∞ 10).

–î–ª—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è —ç—Ç–æ–≥–æ –∑–∞–¥–∞–Ω–∏—è —Å–ª–µ–¥—É–µ—Ç –Ω–∞–ø–∏—Å–∞—Ç—å –ø—Ä–æ–≥—Ä–∞–º–º—É.

```python
text_string = open('files/24.txt').readline()
split_text = text_string.split('A')

max_value = -float('inf')

for a, b, c in zip(split_text, split_text[1:], split_text[2:]):
    if a == b == c:
        max_value = max(max_value,len(c*3))


res = max_value + 4
print(res)
```

- –†–µ–∑—É–ª—å—Ç–∞—Ç (–ø–µ—Ä–≤–∞—è —Å—Ç—Ä–æ–∫–∞ —Ñ–∞–π–ª–∞) —Å–æ—Ö—Ä–∞–Ω—è–µ—Ç—Å—è –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é `text_string`.

### 2. `split_text = text_string.split('A')`

- –ù–∞–ø—Ä–∏–º–µ—Ä, –µ—Å–ª–∏ `text_string = "BACADAEA"`, —Ç–æ `split_text` –±—É–¥–µ—Ç —Ä–∞–≤–µ–Ω `['B', 'C', 'D', 'E']`.
- –†–µ–∑—É–ª—å—Ç–∞—Ç —Å–æ—Ö—Ä–∞–Ω—è–µ—Ç—Å—è –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é `split_text`.

### 3. `print(split_text)`
- –í—ã–≤–æ–¥–∏—Ç —Å–æ–¥–µ—Ä–∂–∏–º–æ–µ —Å–ø–∏—Å–∫–∞ `split_text` –Ω–∞ —ç–∫—Ä–∞–Ω. –≠—Ç–æ –Ω—É–∂–Ω–æ –¥–ª—è –æ—Ç–ª–∞–¥–∫–∏ –∏–ª–∏ –ø—Ä–æ–≤–µ—Ä–∫–∏ –ø—Ä–æ–º–µ–∂—É—Ç–æ—á–Ω–æ–≥–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞.

### 4. `max_value = -float('inf')`

- –≠—Ç–æ –Ω—É–∂–Ω–æ –¥–ª—è —Ç–æ–≥–æ, —á—Ç–æ–±—ã –≤ –¥–∞–ª—å–Ω–µ–π—à–µ–º –º–æ–∂–Ω–æ –±—ã–ª–æ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ —Å—Ä–∞–≤–Ω–∏–≤–∞—Ç—å –¥–ª–∏–Ω—ã —Å—Ç—Ä–æ–∫ –∏ –Ω–∞—Ö–æ–¥–∏—Ç—å –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ.

---

### 5. `for a, b, c in zip(split_text, split_text[1:], split_text[2:]):`
- –ó–¥–µ—Å—å –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è —Ü–∏–∫–ª `for`, –∫–æ—Ç–æ—Ä—ã–π –ø—Ä–æ—Ö–æ–¥–∏—Ç –ø–æ —Ç—Ä–æ–π–∫–∞–º —ç–ª–µ–º–µ–Ω—Ç–æ–≤ —Å–ø–∏—Å–∫–∞ `split_text`.
- `zip(split_text, split_text[1:], split_text[2:])` —Å–æ–∑–¥–∞—ë—Ç –∏—Ç–µ—Ä–∞—Ç–æ—Ä, –∫–æ—Ç–æ—Ä—ã–π –≥—Ä—É–ø–ø–∏—Ä—É–µ—Ç —ç–ª–µ–º–µ–Ω—Ç—ã —Å–ø–∏—Å–∫–∞ –ø–æ —Ç—Ä–∏:
  - `a` ‚Äî —Ç–µ–∫—É—â–∏–π —ç–ª–µ–º–µ–Ω—Ç,
  - `b` ‚Äî —Å–ª–µ–¥—É—é—â–∏–π —ç–ª–µ–º–µ–Ω—Ç,
  - `c` ‚Äî —ç–ª–µ–º–µ–Ω—Ç —á–µ—Ä–µ–∑ –æ–¥–∏–Ω.
- –ù–∞–ø—Ä–∏–º–µ—Ä, –µ—Å–ª–∏ `split_text = ['B', 'C', 'D', 'E']`, —Ç–æ –Ω–∞ –ø–µ—Ä–≤–æ–π –∏—Ç–µ—Ä–∞—Ü–∏–∏:
  - `a = 'B'`, `b = 'C'`, `c = 'D'`.

---

### 6. `if a == b == c:`
- –ü—Ä–æ–≤–µ—Ä—è–µ—Ç—Å—è, —Ä–∞–≤–Ω—ã –ª–∏ –≤—Å–µ —Ç—Ä–∏ —ç–ª–µ–º–µ–Ω—Ç–∞ (`a`, `b`, `c`).
- –ï—Å–ª–∏ –æ–Ω–∏ —Ä–∞–≤–Ω—ã, –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è –±–ª–æ–∫ –∫–æ–¥–∞ –≤–Ω—É—Ç—Ä–∏ —É—Å–ª–æ–≤–∏—è.

### 7. `max_value = max(max_value, len(c * 3))`

- –ó–∞—Ç–µ–º —Å –ø–æ–º–æ—â—å—é —Ñ—É–Ω–∫—Ü–∏–∏ `max` –≤—ã–±–∏—Ä–∞–µ—Ç—Å—è –±–æ–ª—å—à–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –º–µ–∂–¥—É —Ç–µ–∫—É—â–∏–º `max_value` –∏ –¥–ª–∏–Ω–æ–π —ç—Ç–æ–π —Å—Ç—Ä–æ–∫–∏.
- –†–µ–∑—É–ª—å—Ç–∞—Ç —Å–æ—Ö—Ä–∞–Ω—è–µ—Ç—Å—è –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é `max_value`.

- –ü–æ—Å–ª–µ –∑–∞–≤–µ—Ä—à–µ–Ω–∏—è —Ü–∏–∫–ª–∞ –≤—ã—á–∏—Å–ª—è–µ—Ç—Å—è –∏—Ç–æ–≥–æ–≤—ã–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç `res`, –∫–æ—Ç–æ—Ä—ã–π —Ä–∞–≤–µ–Ω `max_value + 4`.
- –ß–∏—Å–ª–æ `4` –¥–æ–±–∞–≤–ª–µ–Ω–æ, –≤–µ—Ä–æ—è—Ç–Ω–æ, –¥–ª—è —É—á—ë—Ç–∞ —Å–∏–º–≤–æ–ª–æ–≤ `'A'`, –∫–æ—Ç–æ—Ä—ã–µ –±—ã–ª–∏ —É–¥–∞–ª–µ–Ω—ã –ø—Ä–∏ —Ä–∞–∑–±–∏–µ–Ω–∏–∏ —Å—Ç—Ä–æ–∫–∏.
