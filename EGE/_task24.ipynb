Текстовый файл состоит не более, чем из 10⁶ символов из набора A, B, C, D, E, F.

Найдите максимальную длину строки вида А*А*А*А, где между буквами А расположены одинаковые группы символов, не содержащие букв А. Например, в строке BDADBADBADBABDAFABDA такая подстрока ADBADBADBA (длина 10).

Для выполнения этого задания следует написать программу.

````python
# исходный код КЭО

text_string = open('files/24.txt').readline()
split_text = text_string.split('A')

max_value = -float('inf')

for a, b, c in zip(split_text, split_text[1:], split_text[2:]):
    if a == b == c:
        max_value = max(max_value,len(c*3))


res = max_value + 4
print(res)
````

### Что делает этот код?
Код ищет в строке `s` подстроку максимальной длины, в которой количество пар "согласная + гласная" не превышает 5. Для этого используется метод "скользящего окна":
- Окно расширяется вправо (`j` увеличивается), если количество пар "согласная + гласная" не превышает 5.
- Окно сдвигается вправо (`i` увеличивается), если количество пар превышает 5.

````python
# Открываем файл и читаем первую строку
with open('24.txt') as file:
    line = file.readline()

# Определяем согласные и гласные буквы
consonants = 'CDF'  # Согласные
vowels = 'AO'       # Гласные

# Инициализируем переменные
max_length = 0      # Максимальная длина подходящей подстроки
pair_count = 0      # Счётчик пар "согласная + гласная"
left = 0            # Левая граница окна
right = 0           # Правая граница окна

# Основной цикл: проходим по строке с помощью "скользящего окна"
while right < len(line):
    # Если количество пар "согласная + гласная" не превышает 5
    if pair_count <= 5:
        # Обновляем максимальную длину подходящей подстроки
        current_length = right - left + 1
        max_length = max(max_length, current_length)

        # Проверяем, есть ли пара "согласная + гласная" на текущей позиции
        if right + 1 < len(line) and line[right] in consonants and line[right + 1] in vowels:
            pair_count += 1  # Увеличиваем счётчик пар

        right += 1  # Расширяем окно вправо

    # Если количество пар превышает 5, сдвигаем левую границу окна
    if pair_count > 5:
        # Если левая граница окна убирает пару "согласная + гласная", уменьшаем счётчик
        if left + 1 < len(line) and line[left] in consonants and line[left + 1] in vowels:
            pair_count -= 1

        left += 1  # Сдвигаем окно вправо

# Выводим результат
print(max_length)
````

````python
# В файле записана последовательность символов.
# Определите в файле максимальное количество идущих подряд цифр.

# Открываем файл для чтения
with open('files/24_01.txt') as file:
    # Читаем первую строку из файла
    text_string = file.readline()

# Переменная для хранения максимального количества подряд идущих цифр
max_count = 0
# Переменная для подсчета текущей последовательности цифр
digits_count = 0

# Строка, содержащая все возможные цифры
digits = '0123456789'

# Проходим по каждому символу в строке
for i in text_string:
    # Проверяем, является ли символ цифрой
    if i in digits:
        # Если символ — цифра, увеличиваем текущий счетчик
        digits_count += 1
        # Обновляем максимальное количество подряд идущих цифр
        max_count = max(max_count, digits_count)
    else:
        # Если символ не цифра, сбрасываем текущий счетчик
        digits_count = 0

# Выводим максимальное количество подряд идущих цифр
print(max_count)
````

````python
# В текстовом файле размером  символов — только заглавные латинские буквы D, R, A, G, O, N. 
# Определите максимальную длину подпоследовательности, 
# в которой каждая из гласных букв (А, О) встречается не более 100 раз.

# Открываем файл и читаем первую строку
with open('files/24 _003.txt') as file:
    text_string = file.readline()

# text_string = 'DDDRRRAAAGGGGOОООDDDRRRAAAGGGGOОООDDDRRRAAAGGGGOОООDDDRRRAAAGGGGOОООDDDRRRAAAGGGGOОООDDDRRRAAAGGGGOООО'

# Определяем согласные и гласные буквы
count_A = 0
count_O = 0

# Инициализируем переменные
max_length = 0      # Максимальная длина подходящей подстроки
left = 0            # Левая граница окна
right = 0           # Правая граница окна

# Основной цикл: проходим по строке с помощью "скользящего окна"
while right < len(text_string):

    if count_A <= 100 and count_O <= 100:

        # Обновляем максимальную длину подходящей подстроки
        current_length = right - left
        max_length = max(max_length, current_length)


        # Увеличиваем счетчики для гласных A и O
        if text_string[right] == 'A':
            count_A += 1
        elif text_string[right] == 'O':
            count_O += 1

        right += 1  # Расширяем окно вправо

    else: 
        # Если количество A или O превышает 100, двигаем left
        if text_string[left] == 'A':
            count_A -= 1
        elif text_string[left] == 'O':
            count_O -= 1
        left += 1  # Сдвигаем окно вправо



# Выводим результат
print(max_length)
````
